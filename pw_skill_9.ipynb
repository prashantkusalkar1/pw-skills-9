Question 7: Name and briefly describe three common evaluation metrics for regression models.

Ans: Regression evaluation metrics measure how well a model’s predictions match actual values. Below are three commonly used metrics, with brief explanations.

1. Mean Absolute Error (MAE)
MAE is the average of the absolute differences between actual and predicted values.

2. Mean Squared Error (MSE)
MSE is the average of the squared differences between actual and predicted values.

3. R-squared (R²)
R² measures the proportion of variance in the dependent variable explained by the model.

Question 8: What is the purpose of the R-squared metric in regression analysis?

Ans: R-squared (R²) is a statistical metric used to measure how well a regression model explains the variability of the dependent variable.

1. Model Fit Assessment

Shows how well the model fits the data

Higher R² → better explanatory power

2. Variance Explanation

Quantifies how much of the outcome’s variation is captured by the predictors

Example: R² = 0.80 means 80% of variance is explained

3. Model Comparison

Helps compare multiple regression models on the same dataset

Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 6, 8, 10])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Get slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Print results
print("Slope (β1):", slope)
print("Intercept (β0):", intercept)
```

Question 10: How do you interpret the coefficients in a simple linear regression model?

Ans: A simple linear regression (SLR) model is written as:

1. Intercept (β₀)
The intercept represents the expected value of Y when X = 0.

It is the baseline level of the dependent variable

Sometimes it has practical meaning, sometimes it is only a mathematical reference


2. Slope (β₁)

The slope represents the average change in Y for a one-unit increase in X, keeping all else constant.

Positive slope → Y increases as X increases

Negative slope → Y decreases as X increases

Magnitude shows strength of impact

Example:
β₁ = 3,000 means salary increases by ₹3,000 for each additional year of experience.

3. Practical Interpretation

Coefficients quantify cause–effect relationships (assuming assumptions hold)

Help in prediction and decision-making
